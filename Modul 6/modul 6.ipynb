```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
# image2 = cv2.imread('tepi.jpg')
# gray_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
# plt.imshow(gray_image2, cmap='gray')

image = Image.open("tepi.jpg")
gray_image = image.convert('L')
plt.imshow(gray_image, cmap='gray')
```

```python
array = np.array(gray_image)
size = gray_image.size
# asd = np.zeros(size, dtype='int')
print(size)
```

sobel

```python
sx = [[-1,0,1],
     [-2,0,2],
     [-1,0,1]]

#konvolusi sx
sobel_x = np.zeros(size, dtype=int)
for i in range(1, size[0]-1):
    for j in range(1, size[1]-1):
        sobel_x[i][j] = array[i-1][j-1]*sx[0][0] + array[i-1][j]*sx[0][1] + array[i-1][j+1]*sx[0][2] + array[i][j-1]*sx[1][0] + array[i][j]*sx[1][1] + array[i][j+1]*sx[1][2] + array[i+1][j]*sx[2][0] + array[i+1][j]*sx[2][1] + array[i+1][j+1]*sx[2][2]
        
print(sobel_x)
```

```python
sy = [[1,2,1],
     [0,0,0],
     [-1,-2,-1]]

#konvolusi sy
sobel_y = np.zeros(size, dtype=int)
for i in range(size[0]-1, size[0]-1):
    for j in range(1, size[1]-115):
        sobel_y[i][j] = array[i-1][j-1]*sy[0][0] + array[i-1][j]*sy[0][1] + array[i-1][j+1]*sy[0][2] + array[i][j-1]*sy[1][0] + array[i][j]*sy[1][1] + array[i][j+1]*sy[1][2] + array[i+1][j]*sy[2][0] + array[i+1][j]*sy[2][1] + array[i+1][j+1]*sy[2][2]
        
print(sobel_y)
```

```python
sobel = abs(sobel_x + sobel_y)
plt.imshow(sobel, cmap='gray')
```

prewit

```python
px = [[-1,0,1],
     [-1,0,1],
     [-1,0,1]]

#konvolusi Px
prewitt_x = np.zeros(size, dtype=int)
for i in range(1, size[0]-1):
    for j in range(1, size[1]-1):
        prewitt_x[i][j] = array[i-1][j-1]*px[0][0] + array[i-1][j]*px[0][1] + array[i-1][j+1]*px[0][2] + array[i][j-1]*px[1][0] + array[i][j]*px[1][1] + array[i][j+1]*px[1][2] + array[i+1][j]*px[2][0] + array[i+1][j]*px[2][1] + array[i+1][j+1]*px[2][2]

print(prewitt_x)
```

```python
py = [[1,1,1],
     [0,0,0],
     [-1,-1,-1]]

#konvolusi Py
prewitt_y = np.zeros(size, dtype=int)
for i in range(1, size[0]-1):
    for j in range(1, size[1]-1):
        prewitt_y[i][j] = array[i-1][j-1]*py[0][0] + array[i-1][j]*py[0][1] + array[i-1][j+1]*py[0][2] + array[i][j-1]*py[1][0] + array[i][j]*py[1][1] + array[i][j+1]*py[1][2] + array[i+1][j]*py[2][0] + array[i+1][j]*py[2][1] + array[i+1][j+1]*py[2][2]

print(prewitt_y)
```

```python
prewit=abs(prewitt_x + prewitt_y)
plt.imshow(prewit, cmap='gray')
```

robert

```python
rx = [[1,0],
     [0,-1]]

#konvolusi rx
robert_x = np.zeros(size, dtype=int)
for i in range(1, size[0]-1):
    for j in range(1, size[1]-1):
        robert_x[i][j] = array[i-1][j-1]*rx[0][0] + array[i-1][j]*rx[0][1] + array[i][j-1]*rx[1][0] + array[i][j]*rx[1][1]

print(robert_x)
```

```python
ry = [[0,1],
     [-1,0]]

#konvolusi ry
robert_y = np.zeros(size, dtype=int)
for i in range(1, size[0]-1):
    for j in range(1, size[1]-1):
        robert_y[i][j] = array[i-1][j-1]*ry[0][0] + array[i-1][j]*ry[0][1] + array[i][j-1]*ry[1][0] + array[i][j]*ry[1][1]

print(robert_y)
```

```python
robert = abs(robert_x + robert_y)
plt.imshow(robert, cmap='gray')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import math
from PIL import Image, ImageFilter
```

```python
image = cv2.imread('noisyimg.png')
new_image = cv2.imread('noisyimg.png')
plt.imshow(image)
prop = image.shape
```

Mean filtering

```python
# 3x3
jmlpad = 1
result = np.zeros(prop)
pad = np.pad(image,(jmlpad,jmlpad),mode='symmetric')
for i in range(jmlpad,len(pad)-jmlpad):
    for j in range(jmlpad,len(pad)-jmlpad): 
        temp = np.ravel(pad[i-jmlpad:i+jmlpad+1,j-jmlpad:j+jmlpad+1])
        result[i-jmlpad,j-jmlpad] = round(np.sum(temp)/len(temp))
plt.imshow(result.astype('uint8'))
```

Modus

```python
def findModus(array):
    temp = np.zeros(int(np.max(array))+1)
    for i in array:
        temp[int(i)]+=1
    max = np.max(temp)
    for i in range(len(temp)):
        if(max == temp[i]): 
            return i

# 3x3
jmlpad = 1
result = np.zeros(prop)
pad = np.pad(image,(jmlpad,jmlpad),mode='symmetric')
for i in range(jmlpad,len(pad)-jmlpad):
    for j in range(jmlpad,len(pad)-jmlpad): 
        temp = np.ravel(pad[i-jmlpad:i+jmlpad+1,j-jmlpad:j+jmlpad+1])
        temp.sort()
        result[i-jmlpad,j-jmlpad] = findModus(temp)
plt.imshow(result.astype('uint8'))
```

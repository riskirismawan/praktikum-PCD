```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('praktikum.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

low = np.array([-0.01,  0.03,  0.03, -0.19, 0.02, 0.63,  0.71,  0.23])
high = np.array([-0.23, 0.71, -0.63, -0.03, 0.19, 0.03, -0.03, -0.01])
```

```python
def konvolusiRow (matriks, kernel):
    padresult = np.pad( matriks, ( (0,0) ,(0,len(kernel)) ), 'constant'  )
    tinggi, lebar = matriks.shape
    convresult = np.zeros([tinggi,lebar])
    for i in range (tinggi ):
        for j in range (lebar):
            for k in range (len(low)):
                convresult[i,j] += kernel[k]*padresult[i,j+k]
                if (convresult[i,j] >255): convresult[i,j] = 255
                elif (convresult[i,j] <0): convresult[i,j] = 0
    return convresult.astype(int)
```

Script di atas merupakan fungsi yang digunakan untuk mengkonvolusikan baris dari matriks citra nantinya dengan kernel yang ada. Hasil arraynya akan di padding sesuai dengan ukuran kernel dengan nilai yang tetap. Neested Loop yang berada dalam fungsi di atas berisikan proses konvolusi dari hasil padding matriks dengan kernel low yang dimana terdapat kondisi hasil yang lebih dari 255 makan akan dianggap 255 dan apabila dibawah 0 akan langsung dianggap 0, hal itu karena rentang rgb hanyalah 0-255. Fungsi tersebut akan mengembalikan nilai hasil konvolusi langsung ke dalam bentuk integer.

```python
def konvolusiColl (matriks, kernel):
    matriks = matriks.transpose()
    padresult = np.pad( matriks, ( (0,0) ,(0,len(kernel)) ), 'constant'  )
    tinggi, lebar = matriks.shape
    convresult = np.zeros([tinggi,lebar])
    for i in range (tinggi ):
        for j in range (lebar):
            for k in range (len(low)):
                convresult[i,j] += kernel[k]*padresult[i,j+k]
                if (convresult[i,j] >255): convresult[i,j] = 255
                elif (convresult[i,j] <0): convresult[i,j] = 0
    convresult = convresult.transpose()
    return convresult.astype(int)
```

Script di atas merupakan fungsi yang digunakan untuk mengkonvolusikan kolom dari matriks citra nantinya dengan kernel yang ada. Sebelumnya perbedaann script ini dengan script fungsi sebelumnya adalah hanya terletak pada proses transpose yang merubah baris menjadi kolom dan sebaliknya agar langsung bisa mendapatkan hasil untuk bagian kolomnya. Hasil arraynya akan di padding sesuai dengan ukuran kernel dengan nilai yang tetap. Neested Loop yang berada dalam fungsi di atas berisikan proses konvolusi dari hasil padding matriks dengan kernel low yang dimana terdapat kondisi hasil yang lebih dari 255 makan akan dianggap 255 dan apabila dibawah 0 akan langsung dianggap 0, hal itu karena rentang rgb hanyalah 0-255. Hasil dari prosesnya akan di transpose. Fungsi tersebut akan mengembalikan nilai hasil konvolusi langsung ke dalam bentuk integer.

```python
def DWTkolom(matriks):
    h,w = matriks.shape
    h2,w2 = int (h/2), int(w/2)
    save = np.zeros ([h,w2])
    for i in range (h):
        for j in range (w2):
            save[i,j] = matriks[i,j*2]
    return save.astype(int)
```

Script di atas merupakan fungsi yang digunakan untuk melakukan transformasi wavelet untuk bagian kolom dari matriks. Proses akan dilakukan sesuai dengan rumus yang nantinya akan ikut merubah bentuk matriks yang akan menampung nilai nilainya. Neested loop tersebut berfungsi agar seluruh bagian kolom dapat ditransformasikan. Fungsi tersebut akan mengembalikan nilai tampungnya ke bentuk integer.

```python
def DWTbaris(matriks):
    h,w = matriks.shape
    h2,w2 = int (h/2), int(w/2)
    save = np.zeros ([h2,w])
    for i in range (h2):
        for j in range (w):
            save[i,j] = matriks[i*2,j]
    return save.astype(int)
```

Script di atas merupakan fungsi yang digunakan untuk melakukan transformasi wavelet untuk bagian baris dari matriks. Proses akan dilakukan sesuai dengan rumus yang nantinya akan ikut merubah bentuk matriks yang akan menampung nilai nilainya. Neested loop tersebut berfungsi agar seluruh bagian baris dapat ditransformasikan. Fungsi tersebut akan mengembalikan nilai tampungnya ke bentuk integer.

```python
def appr (matriks, low, high):
    konvlo1 = konvolusiRow(matriks,low)
    lowdown1 = DWTkolom(konvlo1)
    konvlo2 = konvolusiColl(lowdown1,low)
    lowdown2 = DWTbaris(konvlo2)
    return lowdown2.astype(int)
```

```python
def hori (matriks, low, high):
    konvlo1 = konvolusiRow(matriks,low)
    lowdown1 = DWTkolom(konvlo1)
    konvlo2 = konvolusiColl(lowdown1,high)
    lowdown2 = DWTbaris(konvlo2)
    return lowdown2.astype(int)
```

Script di atas merupakan fungsi yang digunakan untuk pemrosesan nilai array dengan kernelnya secara horizontal pada bentuk array citranya dengan menggunakan fungsi transformasi baris dan kolom yang telah dibuat sebelumnya. Fungsi tersebut akan mengembalikan nilai dengan bentuk integer.

```python
def verti (matriks, low, high):
    konvhi1 = konvolusiRow(matriks,high)
    lowdown1 = DWTkolom(konvhi1)
    konvlo2 = konvolusiColl(lowdown1,low)
    lowdown2 = DWTbaris(konvlo2)
    return lowdown2.astype(int)
```

Script di atas merupakan fungsi yang digunakan untuk pemrosesan nilai array dengan kernelnya secara vertikal pada bentuk array citranya dengan menggunakan fungsi transformasi baris dan kolom yang telah dibuat sebelumnya. Fungsi tersebut akan mengembalikan nilai dengan bentuk integer.

```python
def dia (matriks, low, high):
    konvhi1 = konvolusiRow(matriks,high)
    lowdown1 = DWTkolom(konvhi1)
    konvlo2 = konvolusiColl(lowdown1,high)
    lowdown2 = DWTbaris(konvlo2)
    return lowdown2.astype(int)
```

Script di atas merupakan fungsi yang digunakan untuk pemrosesan nilai array dengan kernelnya secara diagonal pada bentuk array citranya dengan menggunakan fungsi transformasi baris dan kolom yang telah dibuat sebelumnya. Fungsi tersebut akan mengembalikan nilai dengan bentuk integer.

```python
#Hasil untuk level 1
def level(appr, hori, verti, dia):
    result1 = np.concatenate((appr,hori),axis=1)
    result2 = np.concatenate((verti, dia), axis=1)
    result = np.concatenate((result1,result2), axis=0)
    return result.astype(int)
```

Script di atas merupakan fungsi untuk menggabungkan nilai nilai array yang telah di transformasikan pada setiap hasil secara horizontal, vertikal, dan diagonal sesuai rumus agar menjadi matriks yang utuh kembali. Fungsi tersebut akan mengembalikan nilai matriks tersebut dalam bentuk integer.

```python
red   = img[:,:,0]          #untuk mengekstrak warna merah dari img
green = img[:,:,1]          #untuk mengekstrak warna hijau dari img                                            
blue  = img[:,:,2]          #untuk mengekstrak warna biru dari img
#red :
redA, redH, redV, redD = appr(red,low,high), hori(red,low,high), verti(red,low,high), dia(red,low,high)
#green:
greenA, greenH, greenV, greenD = appr(green,low,high), hori(green,low,high), verti(green,low,high), dia(green,low,high)
#blue:
blueA, blueH, blueV, blueD = appr(blue,low,high), hori(blue,low,high), verti(blue,low,high), dia(blue,low,high)
resultred   = level(redA, redH, redV, redD )
resultgreen = level(greenA, greenH, greenV, greenD )
resultblue  = level(blueA, blueH, blueV, blueD )
```

Script di atas berfungsi untuk pemisahan nilai RGB pada citra yang kemudian tiap tiap warna akan di transformasikan dengan transformasi hasil level 1 pada fungsi sebelumnya.

```python
lv1result = np.zeros(img.shape)
lv1result[:,:,0] = resultred.astype(int)
lv1result[:,:,1] = resultgreen.astype(int)
lv1result[:,:,2] = resultblue.astype(int)
```

Script di atas berfungsi untuk menjadikan tiap tiap nilai warna RGB pada citra berbentuk integer.

```python
plt.imshow(lv1result.astype(int))
```

```python
lv2result = lv1result
h,w,r = img.shape
h2,w2 = int (h/2), int(w/2)
LL = np.zeros([h2,w2,r], dtype=int )
LL [:,:,0] = redA
LL [:,:,1] = greenA
LL [:,:,2] = blueA

red2   = LL[:,:,0]
green2 = LL[:,:,1]
blue2  = LL[:,:,2]
```

Script di atas merupakan pemrosesan kembali hasil transformasi level 1 sebelumnya dengan bentuk proses yang sama agar mendapatkan citra dengan transformasi tiap tiap warna level 2.

```python
red2A, red2H, red2V, red2D = appr(red2,low,high), hori(red2,low,high), verti(red2,low,high), dia(red2,low,high)
```

```python
green2A, green2H, green2V, green2D = appr(green2,low,high), hori(green2,low,high), verti(green2,low,high), dia(green2,low,high)
```

```python
blue2A, blue2H, blue2V, blue2D = appr(blue2,low,high), hori(blue2,low,high), verti(blue2,low,high), dia(blue2,low,high)
```

```python
def finalpad(matriks):
    tampung = np.zeros(matriks.shape)
    tampung = np.pad(matriks, ((0,0),(0,1)), 'constant' )
    return tampung
```

Script di atas berfungsi untuk membuat wadah padding yang nantinya akan menampung tiap tiap transformasi warna level 2 agar tertampung pada satu citra pad.

```python
result2red   = level(red2A, red2H, red2V, red2D )
result2green = level(green2A, green2H, green2V, green2D )
result2blue  = level(blue2A, blue2H, blue2V, blue2D )
result2redok = finalpad(result2red)
result2greenok = finalpad(result2green)
result2blueok = finalpad(result2blue)
redLV2   = level(result2redok, redH, redV, redD)
greenLV2 = level(result2greenok, greenH, greenV, greenD)
blueLV2  = level(result2blueok, blueH, blueV, blueD)
lv2 = img
lv2[:,:,0] = redLV2
lv2[:,:,1] = greenLV2
lv2[:,:,2] = blueLV2
```

Script di atas berisikan proses untuk menempatkan hasil transformasi level 2 pada citra yang sebelumnya telah terekstrak. Penempatan tiap warna level 2 ini akan me-replace nilai pada citra awal yang telah di akses sebelumnya, sehingga tiap tiap warna dapat menempati isi dari paddingnya.
